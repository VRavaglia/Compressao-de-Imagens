```python
import numpy as np

def read_p(sub):
    f = open("performances/performances.txt", "r");
    #     vector<vector<performance>> subbs_performances;

    performances = []

    for line in f:
        performance = {}
        split = line.split(',')

        if(split[0] == sub):
#             print(split)
            performance['cbsize'] = int(split[1])
            performance['cbidx'] = int(split[2])
            performance['mse'] = float(split[3])
            performance['R'] = float(split[4])
            performance['bdims'] = [int(split[5]), int(split[6])]
            performances.append(performance)
    return performances
```

```python
bdims = [[1,1],[1,2],[2,1],[2,2],[1,4],[4,1],[2,4],[4,2],[4,4],[1,8],[8,1],[2,8],[8,2],[8,4],[4,8], [8,8], [16, 16]]
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(15, 8), dpi=80)


performances = read_p('1')

for bdim in bdims:
    ps_mse = []
    ps_R = []
    ps_csize = []
    for p in performances:
        if(p['bdims'] == bdim):
            ps_mse.append(p['mse'])
            ps_R.append(p['R'])
            ps_csize.append(p['cbsize'])
    plt.scatter(ps_R, ps_mse, label=str(bdim[0]) + "/" + str(bdim[1]))
    for i in range(0, len(ps_mse)):
        plt.text(ps_R[i], ps_mse[i], str(ps_csize[i]))
plt.legend(fontsize=20)
plt.show()
```
